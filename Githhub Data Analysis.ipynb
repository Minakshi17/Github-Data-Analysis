```
from pyspark.sql import SparkSession 
from pyspark import SparkConf, SparkContext, SQLContext
sc = SparkContext.getOrCreate()
spark = SparkSession(sc)
```

```
df = spark.read.json('FileStore/tables/githubactivity/*_json.gz')
#display(df)
```

```
from pyspark.sql.functions import to_timestamp, hour, col
df_ts= df.withColumn("datetype_timestamp",to_timestamp(col("created_at")))
#df_ts.show()
```

```
data = df_ts.withColumn("hour", hour(col("datetype_timestamp")))
#data.select(data.columns[:]).show()
```

```
# data contains all column and hour colum additionaly
data.createOrReplaceTempView("githubdata")
```

```
payload_data= data.filter("payload.issue.id IS NOT NULL").select("payload.*")
#payload_data.select(payload_data.columns[:]).show()
payload_data.createOrReplaceTempView("payload")
```

##Problem Statements
####1) What type of event is performed the most on this day? 
####2) How many issues were raised per hour on this day?
####3) How many issues were open or closed per hour on this day?

```
%sql
SELECT hour as Time, type as Event, COUNT(type) as No_of_Events FROM githubdata
group by Time, Event 
order by No_of_Events 
```

```

%sql
SELECT hour as Time, type as Event, COUNT(type) as No_of_Events FROM githubdata
where type in('CreateEvent','ForkEvent','PushEvent','DeleteEvent','PullrequestEvent','IssueCommentEvent','IssuesEvent')
group by Time, Event
order by No_of_Events Desc
```

####Problem Statement 2: How many issues were raised per hour on this day?

```
from pyspark.sql.functions import *
issue_count_daily = data.filter("payload.issue.id IS NOT NULL").select("hour").groupBy("hour").agg(count(lit(1)).alias("issue_count")).orderBy('hour')
issue_count_daily.show()
```

```
display(issue_count_daily)
```

Problem 3: How many issues were opened and closed per hour in a day?

```
issues_oc = data.filter("payload.issue.milestone.closed_issues is not null").select(("payload.issue.milestone.*"), "hour")
display(issues_oc)
```

```
issues_oc.createOrReplaceTempView("issues")
```

```
%sql
select hour as time, avg(open_issues) as open_issues, avg(closed_issues) as closed_issues
from issues
group by hour
order by hour
```
